3. Usando "information_schema" en la conexión a base de datos, extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind".

```python
mycursor.execute("""USE information_schema""")

query_customers_columns = """
            SELECT COLUMN_NAME FROM COLUMNS 
            WHERE TABLE_NAME = 'customers' 
            and table_schema = 'northwind'"""
```

```python
mycursor.execute(query_customers_columns)
```

4. Seleccionar todo de la tabla "customers" donde "city" sea "London" o "Seattle". Almacenar los resultados en una variable.

```python
query_customers_city = """ 
                     SELECT *
                     FROM customers
                     WHERE city = 'London' OR city = 'Seattle'"""
mycursor.execute(query_customers_city)
results_customers_city = mycursor.fetchall()
```
